# Categorical Data
Pandas series, tabulated data or a dataframe can be set to the categorical data type.

Categorical data is unordered.   

The parameter ordered, when set to True, creates and ordered categorical.

Pandas supports many operations to b e preformed on categorical data, including:

    - Renaming categories
    - Appending new categories
    - Setting new categories
    - Removing unused categories
    - Sorting and ordering categories

# Text data
    - Read as object, string or text, processing cannot be done in Pandas
    - Data needs to be explicitly input as dtype='string'
