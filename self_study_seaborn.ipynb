Seaborn is a high-level library for preparing statistical graphics.

Closely integrated with Pandas data structures.

Seaborn uses Matplotlib for plotting a variety of graphs.

Features:

    - Set a default theme using set_theme() method on imported sns.
    - Use the load_dataset() method to load data.
    - Inspect data with head() and tail() to view top and bottom parts of the dataset.
    - Use replot() method to plot graphs (provide necessary parameters like data, x, y, hue, style, size, etc.)
    - Provides the displot() method, using which statistical distribution is automatically plotted and visualized.
    - Category plots, Catplot() method to assist category plots, try the parameters to tune the statistical visualization.
    - Provides joinplot() and pairplot() to help visualize and understand complex statistical data.
    - Statistical estimation and error bars.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ADANIPORTS.csv")
df.head()
```

```python
df['H-L'] = df.High - df.Low
```

```python
df['100MA'] = df['Close'].rolling(100).mean() 
```

```python
#Set style to darkgrid
sns.set_style('darkgrid')
```

```python
#Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
z1 = np.linspace(0,10,100)
x1 = np.cos(7*z1) 
y1 = np.sin (7*z1)
```

```python
def return_z(x,y):
    return 50-(x**2+y**2)
```

```python
sns.set_style('whitegrid')
```

```python
x1,y1 = np.linspace(-5,5,50), np.linspace(-5,5,50)
```

```python
x1,y1 = np.meshgrid(x1,y1)
z1 = return_z(x1,y1)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(x1,y1,z1)
plt.show()
```
