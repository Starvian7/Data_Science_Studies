# Data Visualization

Data visualization is the technique to present the data in a pictorial or graphical format.

Benefits:

    - Simplifies the complex quantitative information
    - Analyze and explore Big Data easily
    - Identifies the areas for improvement
    - Identifies ther relationship between data points and variables
    - Explores new patterns and reveals hidden patterns

Python data visualization libraries:
    
    - matplotlib (python 2D plotting library, multi-platform data visualization tool)
    - seaborn
    - vispy 
    - bokeh
    - pygal
    - folium
    - networkx
    

```python
import numpy as np
import matplotlib.pyplot as plt #plot the numbers
from matplotlib import style #set the grid style
%matplotlib inline
```

```python
#generate random numbers
randomNum = np.random.rand(20)
print(randomNum)
```

```python
#select plot style
style.use('ggplot')

#plot the random number
plt.plot(randomNum,'g',label='line one',linewidth=2)

#x axis is number of random numbers (index)
plt.xlabel('Range')

#y axis is actual random number
plt.ylabel('Numbers')

#Title of the plot
plt.title('First Plot')
plt.legend() #plot the graph
plt.show() #display the created plot
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
disaster = 'Volcano Eruption', 'Flood', 'Earthquake', 'Forest Fire', 'Hurricanes', 'Others'
percentile = [62,5,4,2,1,26]
```

```python
#set the pie chart plot properties

#set figure size 
plt.figure(figsize=(10,10))

#explode the largest pie in the data set
explode = (0.05,0,0,0,0,0)

#set pie chart properties
plt.pie(percentile, labels=disaster, explode=explode, autopct='%1.1f%%',startangle=70)

#set axis equal to draw pie as circle
plt.axis('equal')

#set title of the pie chart
plt.title('Disasters')

#show plot
plt.show()
```
