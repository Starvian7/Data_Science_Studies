F-Distribution is Chi-Square of the first sample divided by it's degree of freedom whole 
divided by Chi Square of the second sample divided by degree of freedom of second sample.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import f
```

```python
x = np.linspace(0, 4.5, 1000)
```

```python
data1 = f(1, 1, 0)
data2 = f(5, 8, 0)
data3 = f(4, 4, 0)
data4 = f(100, 200, 0)
```

```python
plt.figure(figsize=(12, 6), dpi = 150)
plt.plot(x, data1.pdf(x), label = '1,1')
plt.plot(x, data2.pdf(x), label = '5,8')
plt.plot(x, data3.pdf(x), label = '4,4')
plt.plot(x, data4.pdf(x), label = '100,200')

plt.legend()
```

Null Hypothesis =  Variances are equal

Alternative Hypothesis are not equal

```python
import random
random.seed(20)
```

```python
x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(100,15) for x in range(20)])
```

```python
f_test_stat = np.var(x, ddof=1) / np.var(y,ddof = 1)
```

```python
dfn = x.size - 1
dfd = y.size - 1
```

```python
p_value = 1 - f.cdf(f_test_stat, dfn, dfd)
```

```python
p_value
```
