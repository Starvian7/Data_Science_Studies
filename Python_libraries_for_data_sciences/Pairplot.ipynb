```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_auto_datasets = pd.read_csv('auto-mpg.csv')
```

```python
df_auto_datasets.head()
```

```python
#write a user defined function for origin 
#1-USA, 2-Europe, 3-Asia
def origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    else:
        return 'Asia'

#use apply function
df_auto_datasets['origin'] = df_auto_datasets['origin'].apply(origin)
```

```python
df_auto_datasets.head(30)
```

```python
# draw the pair plot using sns for mpg, weight, origin and with hue origin, set the size to 4
# note: hue is variable in data set to map plot aspects to different colors
sns.pairplot(df_auto_datasets[['mpg', 'weight', 'origin']], hue = 'origin', height = 4)
```
