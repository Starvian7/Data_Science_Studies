```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_1 = np.random.chisquare(df = 1, size = 1000)
data_2 = np.random.chisquare(df = 2, size = 1000)
data_3 = np.random.chisquare(df = 3, size = 1000)

print(data_1[:10])
```

```python
sns.distplot(data_1,hist = False, label = 'dof 1')
sns.distplot(data_2,hist = False, label = 'dof 2')
sns.distplot(data_3,hist = False, label = 'dof 3')
plt.legend()
```

```python
#Two-samples
from scipy.stats import chi2_contingency
```

```python
data = [[10, 20, 30], [6, 9, 17]]
```

```python
stat, p_value, dof, chi_array = chi2_contingency(data)
```

```python
p_value
```

```python
data = [[10, 20, 30], [9, 1, 8]]
```

```python
stat, p_value, dof, chi_array = chi2_contingency(data)
```

```python
p_value #hypothesis accepted as there is no association between the groups that were tested.
```
