# T-Test
## When is it used:
    - When we want to determine if means of 2 groups are different or not
    - When we want to compare means of two samples
    - Sample size <=30
    - Data follows normal distribution

# Independent Sample T-Test:
    1. Check the average of 2 independent unrelated groups.
    2. Samples should be from 2 different populations.
    
    - Null Hypothesis : μa = μb
    - Alternate Hypothesis : μa != μb

# Paired Sample T-Test:
    1. Average of 2 samples taken from the same population but in different points in time.
    
    - Null Hypothesis : μd = 0
    - Alternate Hypothesis : μd != 0

# One Sample T-Test:
    1. Average of a single group is different from known average.
    
    - Null Hypothesis: μa = X
    - Alternate Hypothesis: μa != X

```python
import random
import seaborn as sns
```

```python
a = [random.gauss(50,20) for x in range(30)]
b = [random.gauss(55,15) for x in range(30)]
```

```python
sns.set_style('darkgrid')
sns.kdeplot(a,fill = True)
sns.kdeplot(b,fill = True)
```

```python
import scipy.stats as stats
import numpy as np
```

```python
t_stat, p_value = stats.ttest_ind(a, b, equal_var = False) #independent sample T-Test
```

```python
t_stat, p_value
```

```python
np.mean(a), np.mean(b) 
```

```python
a = [random.gauss(50,20) for x in range(30)]
b = [random.gauss(60,25) for x in range(30)]
```

```python
t_stat, p_value = stats.ttest_rel(a,b) #paired sample T-Test
```

```python
t_stat, p_value
```

```python
np.mean(b)-np.mean(a) #difference is not 0, hence rejects the null hypothesis
```

```python
np.mean(a)
```

```python
t_stat, p_value = stats.ttest_1samp(a, 40, axis = 0) #one sample T-Test
```

```python
t_stat, p_value
```

# One Sample Z-Test

```python
import random
random.seed(20)
from statsmodels.stats.weightstats import ztest as ztest
```

mean = 100

standard deviation = 15

```python
a = [random.gauss(100,15) for x in range(40)]
```

```python
ztest(a, value = 100)
```

```python
a = [random.gauss(120,15) for x in range(40)]
```

```python
ztest(a, value = 100, alternative = 'larger')
```

```python
a = [random.gauss(80,15) for x in range(40)]
```

```python
ztest(a, value = 100, alternative = 'smaller')
```

# Two Sample Z-Test

```python
a = [random.gauss(100,15) for x in range(40)]
b = [random.gauss(120,15) for x in range(40)]
```

```python
ztest(a, b, value = 0)
```

```python
a = [random.gauss(100,15) for x in range(40)]
b = [random.gauss(100,15) for x in range(40)]
```

```python
ztest(a, b, value = 0) #accepted the null hypothesis
```
