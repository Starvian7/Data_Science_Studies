```python
#Histogram and Scatter plots
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import style

%matplotlib inline
```

```python
#Load wine dataset
loadWineData = load_wine()
```

```python
#view wine dataset

data=pd.DataFrame(data=np.c_[loadWineData['data'],loadWineData['target']],columns=loadWineData['feature_names']+['target'])

data.info()
```

```python
#difine x axis for the data
x_axis = loadWineData.data
```

```python
#define y axis for the target
y_axis = loadWineData.target
```

```python
#plot histogram
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y_axis,bins=50)
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

```python
#plot scatter plot
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(loadWineData.data[:,5],loadWineData.target)
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

```python
#Pie Chart
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#job data in percentile
jobData = ['40','20','17','8','5','10']

#define label as different departments
labels = 'IT', 'Finance', 'Marketing', 'Admin', ' HR', 'Operations'

#explode the 1st slice which is IT
explode = (0.07, 0, 0, 0, 0, 0)

#draw the pie chart and set the parameters
plt.pie(jobData, labels=labels, explode=explode)

#show the plot
plt.show()
```
